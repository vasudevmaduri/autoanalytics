```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

URL_home = "https://www.shopee365.com"
page = requests.get(URL_home)

data=[]
```

```python

soup = BeautifulSoup(page.content, 'html.parser')
job_elems = soup.find("ul",{'class':'groupmenu'})
```

```python

links = [a['href'] for a in job_elems.find_all('a', href=True)]
links = list(filter(None, links))
links = links[:len(links)-3]
hash_ = [i for i, x in enumerate(links) if x == "#"]

for i in hash_:
    del links[links.index("#")]
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

URL = "https://www.shopee365.com/car-door-outer-handles?p=10"
page = requests.get(URL)

data=[]
```

```python

import time
start_time =time.time()
spares_df = pd.DataFrame()
spare_type=['car-door-outer-handles']

for spare in links[links.index("https://www.shopee365.com/car-control-arm"):]:
    for i in range(1,6):
        temp_df = pd.DataFrame()
        print("Running {} - Page {}".format(spare, i))
        market_type, brand, product, new_price_, old_price_, query_string_, page_id_ =[],[],[],[],[],[],[]
        URL = spare+"?p="+str(i)
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, 'html.parser')
        try:
            job_elems = soup.find("ol",{'class':'products list items product-items same-height'})
            for item in job_elems.findAll('li'):
                type_ = item.find_all("div",class_= 'cls-manufacturer-listing')[0].text
                model_name = item.find_all("div", class_ = 'cls-vehicle-make-listing')[0].find_all('span')[0].text
                list_tags = item.find_all('a')
                list_span = item.find_all('span')
                spare_name = list_span[0].find('img',alt=True).attrs['alt']
                new_price = item.find_all("span", class_="price")[0].text[1:]
                old_price = item.find_all("span", class_="price")[1].text[1:]
    #             print(spare,i, type_, model_name, spare_name, new_price, old_price)
    #             print(new_price, old_price)
                query_string_.append(spare)
                page_id_.append(i)
                market_type.append(type_)
                brand.append(model_name)
                product.append(spare_name)
                new_price_.append(str(new_price))
                old_price_.append(str(old_price))
    #             print("------")
        except:
            print("Error")
        temp_df["query_string"] = query_string_
        temp_df["page_id"]=page_id_
        temp_df["market_type"]=market_type
        temp_df["brand"]=brand
        temp_df["product"]=product
        temp_df["new_price_"]=new_price_
        temp_df["old_price_"]=old_price_
        spares_df = spares_df.append(temp_df)
        time.sleep(2)
print((time.time() - start_time))
```

```python
spares_df.to_csv("spares_df_v4.csv")
```
